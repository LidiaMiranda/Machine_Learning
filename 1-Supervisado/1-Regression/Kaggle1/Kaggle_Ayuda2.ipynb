## Data processing

```python
#Convertir marcas en números en base a su value count. Creamos diccionario
marcas = {'Lenovo': 1, 'Dell': 2, 'HP': 3, 'Asus': 4, 'Acer': 5, 'MSI': 6, 'Toshiba': 7}
#Reemplazamos los valores de marcas
train["Company"] = train["Company"].replace(marcas)
```

```python
#Convertir tipos en números. Creamos diccionario
tipos = {'Notebook': 1, 'Gaming': 2, 'Ultrabook': 3, '2 in 1 Convertible': 4}
#Reemplazamos los valores de tipos
train["TypeName"] = train["TypeName"].replace(tipos)
```

```python
#Tratamos la columna ScreenResolution para dividirla en 3. Creamos la columna ScreenInfo para obtener la información de la calidad y si es táctil
train["ScreenInfo"] = train["ScreenResolution"].str[:-9]
train['ScreenResolution'].fillna('Unknown', inplace=True)
train["ScreenInfo"].unique()
```

```python
# Crear la columna "ScreenQuality"
train['ScreenQuality'] = train['ScreenInfo'].str.replace('Touchscreen', '').str.strip()

# Crear la columna "Touchscreen"
train['Touchscreen'] = train['ScreenInfo'].str.contains('Touchscreen').astype(int)

#Cambiar columna "ScreenResolution" a sólo la resolución
train['ScreenResolution'] = train['ScreenResolution'].str[-9:]
```

```python
#Convertimos ScreenQuality a numérica
calidad = {'':0,'IPS Panel Full HD': 1,'IPS Panel Full HD /': 1, 'Full HD':1, "Full HD /":1}
train["ScreenQuality"] = train["ScreenQuality"].replace(calidad)
```

```python
#Convertimos ScreenResolution a numérica ordenado por value counts
resolucion = {'1920x1080':1, '1366x768':2}
train["ScreenResolution"] = train["ScreenResolution"].replace(resolucion)
```

```python
#Quitar los GHz de la CPU
train['Cpu'] = train['Cpu'].str[:-7]
#Separar cada palabra en una columna
CPU =train["Cpu"].str.split(expand = True)
```

```python
#Todos son Intel Core
# Crear la columna "Cpu_ICore"
train["Cpu_ICore"] = train["Cpu"].str.split().str[2:].apply(lambda x: ' '.join(filter(None, x)))
```

```python
#Creamos un diccionario con los modelos de CPU y los pasamos a número
intel_core = {'i5 7200U': 1,
               'i7 7700HQ': 2,
               'i3 600': 3,
               'i5 6200U': 4,
               'i5 8250U': 5,
               'i7 7500U': 6,
               'i3 7100U': 7,
               'i7 8550U': 8,
               'i7 6500U': 9,
               'i7 6700HQ': 10,}
train["Cpu"] = train["Cpu_ICore"].replace(intel_core)
```

```python
#Eliminamos GB de RAM
train["Ram"] = train["Ram"].str.replace("GB", "")
train['Ram'] = train['Ram'].astype(int)
```

```python
#Nos quedamos solo con la memoria en GB y lo pasamos a int
train["Memory"] = train["Memory"].str.replace('^128GB SSD +.*', '1128')
train["Memory"] = train["Memory"].str.replace('^256GB SSD +.*', '2256')
train["Memory"] = train["Memory"].str.split("G.*").str[0]
train["Memory"] = train["Memory"].str.replace('^1TB.*', '1000')
train["Memory"] = train["Memory"].astype(int)
```

```python
#Convertimos los valores de las tarjetas gráficas a números

graficas = {'Nvidia GeForce GTX 1050 Ti': 1,
        'Intel HD Graphics 620': 2,
        'Intel HD Graphics 520': 3,
        'Intel UHD Graphics 620': 4,
        'Nvidia GeForce GTX 1060': 5,
        'AMD Radeon 530': 6,
        'Nvidia GeForce 940MX': 7,
        'AMD Radeon R5 M430': 8,
        'Nvidia GeForce GTX 1050': 9,
        'Nvidia GeForce GTX 1070': 10}
train["Gpu"] = train["Gpu"].replace(graficas)
```

```python
#En la columna "OpSys", convertimos los sistemas en números. Creamos diccionario
sistema = {'No OS': 0, 'Windows 10': 1, 'Linux': 2,'Windows 7': 3}
train["OpSys"] = train["OpSys"].replace(sistema)
```

```python
#En la columna "Weight", eliminar kg y convertirlo a float.
train["Weight"] = train["Weight"].str.replace("kg", "")
train["Weight"] = train["Weight"].astype(float)
```

```python
#Eliminamos las columnas "ScreenInfo" y "Cpu_ICore"
train = train.drop(labels=["ScreenInfo", "Cpu_ICore" ], axis=1)
train
```
